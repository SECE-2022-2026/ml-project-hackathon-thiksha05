```python
import pandas as pd
import numpy as np
```

```python
true = pd.read_csv('True.csv')
fake = pd.read_csv('Fake.csv')
```

```python
true['label']=0
fake['label']=1
```

```python
fake.head()
```

```python
news = pd.concat([true, fake])
```

```python
news.isnull().sum()
```

```python
news = news.drop(['title', 'subject', 'date'], axis=1)
```

```python
news = news.sample(frac=1)   #random shuffle the rows of the data set
```

```python
news.head()
```

```python
news.reset_index(inplace=True)
```

```python
news.head()
```

```python
news.drop(['index'],axis=1, inplace=True)
```

```python
import re
```

```python
#function for text preprocessing

def wordopt(text):
  #convert into lowercae
  text=text.lower()
  #remove urls

  text = re.sub(r'https?://\S+|www\.\S+','',text)

  #remove html tags
  text = re.sub(r'<.*?>','',text)

   #remove punctuation
  text = re.sub(r'[^\w\s]','',text)

   #remove digits
  text = re.sub(r'\d+','',text)
  #remove newline character
  text = re.sub(r'\n','',text)

  return text





```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
news.head()
```

```python
x = news['text']
y = news['label']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vetorization = TfidfVectorizer()
xv_train = vetorization.fit_transform(x_train)
xv_test = vetorization.transform(x_test)
```

```python
xv_train
```

```python
xv_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR = LogisticRegression()
LR.fit(xv_train,y_train)
```

```python
pred_lr=LR.predict(xv_test)
```

```python
LR.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import  DecisionTreeClassifier
```

```python
DTC = DecisionTreeClassifier()
```

```python
DTC.fit(xv_train,y_train)
```

```python
pred_dtc = DTC.predict(xv_test)
```

```python
DTC.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_dtc))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(xv_train,y_train)
```

```python
predict_rfc = rfc.predict(xv_test)
```

```python
rfc.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predict_rfc))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
gbc = GradientBoostingClassifier()
```

```python
gbc.fit(xv_train,y_train)
```

```python
pred_gbc = gbc.predict(xv_test)
```

```python
gbc.score(xv_test, y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_gbc))
```

```python
def output_lable(n):
    if n == 0:
        return "fake news"
    elif n == 1:
        return "true News"
```

```python
def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vetorization.transform(new_x_test)
    pred_lr = LR.predict(new_xv_test)
    pred_gbc = gbc.predict(new_xv_test)
    pred_rfc = rfc.predict(new_xv_test)
    return "\n\nLR Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_lable(pred_lr[0]),
                                                                                       output_lable(pred_gbc[0]),
                                                                                       output_lable(pred_rfc[0]))
```

```python
news_article = str(input())
```

```python
manual_testing(news_article)
```
